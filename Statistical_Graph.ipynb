```python
#matplotlib
import matplotlib. pyplot as plt
import pandas as pd
import numpy as np
```

```python
Janvi=[1,2,3,4,5,]
placement=[10,20,30,40,50]
plt.bar(Janvi, placement, color='red')
plt.show()
```

```python
categories = ["A","B", "C", "D", "E"] #x axis 
values  = [5,6,7,8,9] #y axis
widths=[0.7,0.5,0.8,0.6,0.5] #widths hots hai gap ke liye
plt.bar(categories, values, color='skyblue',edgecolor='black', width=widths)
plt.show
```

```python
categories = ["A","B", "C", "D", "E"] #x axis 
values  = [5,6,7,8,9] #y axis
widths=[0.7,0.5,0.8,0.6,0.5] #widths hots hai gap ke liye
plt.bar(categories, values, color='skyblue',edgecolor='black', width=widths, label='stts result')
plt.xlabel('student')
plt.ylabel('pass marks')
plt.title('stts sub')
plt.grid(axis='y',linestyle='dashdot', alpha=0.9) #alpha is size of line or axis hmmchange kar sakte hai 
plt.legend()
plt.show()
```

```python
#horizontal graph (plt.bar ke samnee h hoga bas like plt.barh)
categories = ["A","B", "C", "D", "E"] #x axis 
values  = [5,6,7,8,9] #y axis
widths=[0.7,0.5,0.8,0.6,0.9] #widths hots hai gap ke liye
plt.barh(categories, values, color='skyblue',edgecolor='black')
plt.xlabel('student')
plt.ylabel('pass marks')
plt.title('stts sub')
plt.grid(axis='y',linestyle='dashdot', alpha=0.9) #alpha is size of line or axis hmmchange kar sakte hai 
plt.show()
```

```python
# line chart or line margene(line chart mai grib be plt.plot ke under he aata hai or ye plt.plot ke name se descover hota hai)
x = [1,2,3,4,5,6]
y = [10,20,30,40,50,60]
plt.plot(x,y, color='red', linestyle='dashed',marker='o', label='stts result') #label ak he hota hai or kissi be side mai la kae ja sakte hai or label lagend ke name se show hoga 
plt.xlabel('Student')  
plt.ylabel('Stats Mark')
plt.title('performance')
plt.legend()
plt.show()
```

```python
x = [1,2,3,4,5,6]
y = [10,20,30,40,50,60]
plt.plot(x,y, color='red', linestyle='dashed',marker='o', label='stts result') #label ak he hota hai or kissi be side mai la kae ja sakte hai or label lagend ke name se show hoga 
plt.xlabel('Student')  
plt.ylabel('Stats Mark')
plt.title('performance')
plt.legend() #legand label btata hai
plt.grid(axis='y') #grid ko hmm change be kar sakte hai (like axis= 'x', ya fir axis = 'y' or agr 2 mai karni ho toh plt.grid()bas.)
plt.show()
```

```python
#scatter chart 
x = [1,2,3,4,5,6, 7, 8, 9, 12, 13, 14]
y = [10,20,30,40,50,60, 68,90, 45, 56, 54,68]
plt.scatter(x,y, color="r",linestyle="dashed", marker="*", label="y=x^2")  #marker change kar sakte hai linestyle color legend 
plt.xlabel("Student")  
plt.ylabel("Stats Mark")
plt.title("Scatter plot") #Title de sakte hai 
plt.grid()
plt.legend() # label batata hai 
plt.show()
```

```python
x = [1,2,3,4,5,6, 7, 8, 9, 12, 13, 14]
y = [10,20,30,40,50,60, 68,90, 45, 56, 54,68]
size = [100,20,30,60,500,65,56,300,150,200,300,357]
plt.scatter(x,y, color='red', s=size, linestyle='dashed', marker='*', label='stts result')
plt.xlabel("Student")  
plt.ylabel("Stats Mark")
plt.title("Scatter plot")
plt.grid(True, linestyle="dashed", alpha=0.8) #scatter mai grid mai axis nhi True likhna hai or true ka t captial hoga
plt.legend() # label batata hai 
plt.show()
```

```python
#histogram(histogram plt.hist)
data = np.random.randn(100)
plt.hist(data, bins=25, color="r", alpha=1, edgecolor="black", label="mark") #bin graph size 
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Basic Histogram")
plt.grid()
plt.legend(loc=1)
plt.show()
```

```python
data = np.random.randn(100)
np.random.seed(40) #Fix karta hai random number ko 
plt.hist(data, bins=25, color="r", alpha=1, edgecolor="black", label="mark") #bin graph size 
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Basic Histogram")
plt.grid(axis='y')
plt.legend(loc=1)
plt.show()
```

```python
sizes = [40, 50, 60, 70]
labels = ["Group1", "Group2", "Group3", "Group4"]
colors = ["blue", "green", "yellow", "orange"]

plt.pie(sizes, labels=labels, autopct="%1.1f%%", startangle=140, colors=colors)
plt.title("Pie chart Example")
plt.show()
```

```python
sizes=[40,50,60,70]
labels=["Group1", "Group2", "Group3", "Group4"]
colors=["blue", "green", "yellow", "orange"]
plt.pie(sizes, labels=labels, autopct="%1.1f%%", startangle=140, colors=colors, explode=(0.3,0,0,0))  
plt.title("Pie chart Example")
plt.show()
```

```python
w=0.3 #width
x= ["C1", "C2", "C3", "C4"]
boys=[100,40,50,30]
girls=[200,70,80,50]
bar1=np.arange(len(x)) #kitne bar banage 
plt.bar(bar1,boys,w,label="boys")
bar2=[i+w for i in bar1]
plt.bar(bar2, girls, w, label="girl")
plt.xticks(bar1+w/2,x)
plt.show()
```
